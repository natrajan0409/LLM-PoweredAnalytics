```python
from mistralai.client import MistralClient
from mistralai.models.chat_completion import ChatMessage  # Correct import for v2.3.2
from dotenv import load_dotenv
import os

# Load API key
load_dotenv()
api_key = os.getenv("oNYGwY98i3QylaJNPON70ucpg1s4RAiF")  # Ensure .env has: MISTRAL_API_KEY=your_key_here

# Initialize client
client = MistralClient(api_key=api_key)

# Define messages (using ChatMessage for clarity)
messages = [
    ChatMessage(role="user", content="Hello Mistral!")
]

# Send chat request
response = client.chat(
    model="mistral-small",  # Start with a smaller model
    messages=messages
)

# Print response
print(response.choices[0].message.content)
```
